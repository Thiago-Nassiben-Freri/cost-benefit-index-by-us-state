```python
import pandas as pd


# Transform a csv file into a pandas dataframe

df = pd.read_csv("datasets\house_data.csv")

df
```

```python
# Get the overall average home price in the United States 

average_house_price_us = df["median_home_value"].mean()
average_house_price_us = f"US$ {average_house_price_us:,.2f}"

print(average_house_price_us)
```

```python
# Get the overall average outstanding mortgage balance in the United States

average_out_mortgage_balance = df["average_outstanding_mortgage_balance"].mean()
average_out_mortgage_balance_us = f"US$ {average_out_mortgage_balance:,.2f}"

print(average_out_mortgage_balance_us)
```

```python
# Get the overall average fico score in the United States

average_fico_score = df["average_fico_score"].mean()
average_fico_score_us = f"{average_fico_score:.2f}"

print(average_fico_score_us)
```

```python
states_house_price = df.groupby("state")["median_home_value"].max()
top_5 = states_house_price.nlargest(5)
i = 0

for state, val in top_5.items():
    print(f"{i + 1} - {state:<25} : US${val:,.2f}")
    i +=1
```

```python
least_5 = states_house_price.nsmallest(5).sort_values(ascending=False)
j = len(df["state"]) - 5

for state, val in least_5.items():
    print(f"{j + 1} - {state:<25} : US${val:,.2f}")
    j += 1
```

```python
fico_score_by_state = df.groupby("state")["average_fico_score"].max()
top_5_fico = fico_score_by_state.nlargest(5)
y = 0

for state, fico in top_5_fico.items():
    print(f"{y + 1} - {state:<25} : {fico}/850")
    y += 1
```

```python
least_5 = fico_score_by_state.nsmallest(5).sort_values(ascending=False)
z = len(df["state"]) - 5

for state, fico in least_5.items():
    print(f"{z + 1} - {state:<25} : {fico}/850")
    z += 1
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.stats import linregress


# Figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Variables
house_value = df["median_home_value"] / 1_000
income = df["income_per_capita"] / 1_000

# Calculate a linear regression
slope, intercept, r_value, p_value, std_err = linregress(income, house_value)
regression_line = slope * income + intercept

# Ploting 
plt.scatter(income, house_value)
plt.plot(income, regression_line, color='red')

# Axis labels
ax.set_ylabel("House Value", weight="bold", loc="top")
ax.set_xlabel("Income", weight="bold", loc="left")

# Limit of axis y and x
ax.set_ylim(100, 950)
ax.set_xlim(40, 120)

# Format the axis to "US$"
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'US$ {x:,.0f}K'))
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'US$ {x:,.0f}K'))

# Remove the borders
ax.spines[["top", "right"]].set_visible(False)

# Show 
plt.show()
```

```python
df_4 = pd.read_csv("datasets\life_exp.csv")

df_4
```

```python
df["life_expectancy"] = df_4["Mean"]

df
```

```python
df[["median_home_value", "crime_rate"]].corr()
```

```python
norm_value = (df["median_home_value"].max() - df["median_home_value"]) / (df["median_home_value"].max() - df["median_home_value"].min())
norm_crime_rate = (df["crime_rate"].max() - df["crime_rate"]) / (df["crime_rate"].max() - df["crime_rate"].min())
norm_mortage = (df["average_outstanding_mortgage_balance"].max() - df["average_outstanding_mortgage_balance"]) / (df["average_outstanding_mortgage_balance"].max() - df["average_outstanding_mortgage_balance"].min())
norm_income = (df["income_per_capita"].max() - df["income_per_capita"]) / (df["income_per_capita"].max() - df["income_per_capita"].min())
norm_life_exp = (df["life_expectancy"].max() - df["life_expectancy"]) / (df["life_expectancy"].max() - df["life_expectancy"].min())
```

```python
cost_benefit_index = (norm_value * 0.2 + norm_crime_rate * 0.2 + norm_mortage * 0.2 + norm_income * 0.2 + norm_life_exp * 0.2)

cost_benefit_index
```

```python
df["cost_benefit_index"] = cost_benefit_index

df
```

```python
cost_benefit_idx_state = df.groupby("state")["cost_benefit_index"].max()
top_cost_benefit_idx_state = cost_benefit_idx_state.nlargest(5)
a = 0

for state, index in top_cost_benefit_idx_state.items():
    print(f"{a + 1} - {state} : {index:,.2f}/1")
    a += 1
```

```python
cost_benefit_idx_state = df.groupby("state")["cost_benefit_index"].min()
least_cost_benefit_idx_state = cost_benefit_idx_state.nsmallest(5).sort_values(ascending=False)
a = 46

for state, income in least_cost_benefit_idx_state.items():
    print(f"{a + 1} - {state} : {income:,.2f}/1")
    a += 1
```
